**Data Cleaning**

```python
data.isnull()
```

```python
#Finding out null values in dataset
data.isnull().sum()/len(data)*100
```

```python
data["Region_Code"]=data["Region_Code"].astype("int64")
data["Annual_Premium"]=data["Annual_Premium"].astype("int64")
```

```python
data['Policy_Sales_Channel']=data['Policy_Sales_Channel'].astype("int64")
```

```python
#removing unimportant column
data=data.drop("Driving_License",axis=1)
```

```python
data["Vehicle_Age"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
a=["Gender","Vehicle_Age","Vehicle_Damage"]
for i in a:
  data[i]=le.fit_transform(data[i])

data.head()
```

```python
#correlation for given dataset
data.corr()
```

```python
#heatmap for the given correlation
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(),annot=True)
plt.title("HEATMAP", fontweight="bold")
plt.show()
```

**Data Visualization**

```python
#Analysing the age distribution
plt.figure(figsize=(10,8))
sns.histplot(data["Age"],color="pink", kde=True, bins=30)
plt.title("Age Distribution of Customers", fontsize=15, fontweight="bold")
plt.xlabel("Age",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.show()
```

```python
#Analysing the age distribution and impact on insurance claims
plt.figure(figsize=(10,8))
sns.boxplot(x="Response",y="Age",color="pink",data=data)
plt.title("Age Impact On Insurance Claims", fontsize=20, fontweight="bold")
plt.xlabel("Insurance Claims",fontsize=20)
plt.ylabel("Age",fontsize=20)
plt.show()
```

```python
#Investigating the distribution of insurance premiums.
plt.figure(figsize=(10,8))
sns.histplot(data["Annual_Premium"], kde=True, bins=30, color="pink")
plt.title("Annual Premium Distribution", fontsize=20, fontweight="bold")
plt.xlabel("Premium Amount",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.show()
```

```python
#Investigating the distribution of insurance premiums vs claim frequencies.
plt.figure(figsize=(10,8))
sns.boxplot(x="Response",y='Annual_Premium',color="pink",data=data)
plt.title("Annual Premium vs Insurance Claim", fontsize=20, fontweight="bold")
plt.xlabel("Insurance Claim",fontsize=20)
plt.ylabel("Annual_Premium",fontsize=20)
plt.show()
```

Exploring factors contributing to higher claim frequencies

```python
plt.figure(figsize=(10,8))
sns.barplot(x="Vehicle_Damage",y="Response",data=data,color='pink')
plt.xlabel("Vehicle_Damage",fontsize=20)
plt.ylabel("Insurance claim",fontsize=20)
plt.show()
```

 If the customer’s vehicle was previously damaged, they are much more likely to opt for insurance.Therefore it has positive correlation.

```python
plt.figure(figsize=(10,8))
sns.barplot(x="Previously_Insured",y="Response",data=data,color='pink')
plt.xlabel("Previously_Insured",fontsize=20)
plt.ylabel("Insurance claim",fontsize=20)
plt.show()
```

Customers who were previously insured are far less likely to respond again.

```python
plt.figure(figsize=(10,8))
sns.barplot(x="Vehicle_Age",y="Response",data=data,color='pink')
plt.xlabel("Vehicle_Age",fontsize=20)
plt.ylabel("Insurance claim",fontsize=20)
plt.show()
```

Customers with older vehicles show higher interest in insurance claims.

Investigate the role of gender in insurance claims

```python
#Investigate the role of gender in insurance claims
plt.figure(figsize=(10,8))
sns.countplot(x='Gender', hue='Response',color="pink", data=data)
plt.title("Insurance Response WRT Gender",fontsize=20, fontweight="bold")
plt.xticks([0, 1], ['Female', 'Male'],fontsize=20)
plt.xlabel("Gender",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.show()
```

females have a marginally higher response rate , but in numbers males are ahead of females .

```python
#the impact of vehicle age on the likelihood of a claim
plt.figure(figsize=(10,8))
sns.countplot(x='Vehicle_Age', hue='Response',color='pink', data=data)
plt.title("Insurance Claim WRT Vehicle_Age",fontsize=20, fontweight="bold")
plt.xlabel("Vehicle Age",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.show()
```

```python
#regional patterns in insurance claims
plt.figure(figsize=(10,8))
sns.barplot(x='Region_Code',y='Response',data=data,color='pink')
plt.title("Regional Claim Likelihood")
plt.xlabel("Region Code")
plt.ylabel("Insurance Claim")
plt.xticks(rotation=90)
plt.show()
```

```python
#impact of different insurance policy types
plt.figure(figsize=(10,8))
policy_sales_channel = data.groupby('Policy_Sales_Channel')['Response'].mean().head(15)
policy_sales_channel.plot(kind='bar', color='pink')
plt.title("Policy Sales Channels by Insurance Claim",fontsize=20,fontweight="bold")
plt.xlabel("Policy Sales Channel")
plt.ylabel("Insurance Claim")
plt.show()
```

Few policy channels dominate in insurance claims while others contribute less .

```python
#relationship between vehicle damage and claim frequencies
plt.figure(figsize=(10,8))
sns.barplot(x='Vehicle_Damage',y='Response',data=data,color='pink')
plt.title("Vehicle Damage vs Insurance Claim Response",fontsize=20,fontweight="bold")
plt.xlabel("Vehicle Damage")
plt.ylabel("Response")
plt.show()
```

```python
#the number of policies held by a customer influences claim likelihood.
plt.figure(figsize=(10,8))
sns.histplot(data["Vintage"], bins=30, kde=True,color="pink")
plt.title("Customer Loyalty",fontsize=20,fontweight="bold")
plt.xlabel("Vintage")
plt.show()
```

Customers with longer tenure appear more likely to respond, indicating loyalty impact.

```python
#temporal patterns in insurance claims
plt.figure(figsize=(10,8))
sns.boxplot(x="Response", y="Vintage",color="pink", data=data)
plt.title("Vintage vs Insurance Response",fontsize=20)
plt.xlabel("Insurance Claim")
plt.ylabel("Vintage")
plt.show()
```

## ✅ **Key Summary Insights and Recommendations**

---

### 1. 🚗 **Vehicle Damage is a Strong Indicator of Insurance Response**

- Customers whose vehicles had **previous damage** are **much more likely** to respond positively to insurance offers.  
-  **Targeting this segment** can significantly increase conversions and profitability.

---

### 2. 🕒 **Vehicle Age Strongly Influences Interest**

- Owners of **older vehicles** (especially those older than 2 years) show **higher interest** in insurance.  
-  This may be due to higher perceived risk and potential repair costs.  
-  Focus campaigns on this group for better ROI.

---

### 3. 👵 **Age Factor vs Insurance Interest**

- Customers in the **45+ age group** display a **slightly higher tendency** to opt for insurance coverage.  
-  They likely value **financial security** and **risk protection** more than younger demographics.  
-  Customize marketing strategies to appeal to this mature age segment.

---

### 4. 📈 **Sales Channel’s Influence on Response Rate**

- A few **specific sales channels** are responsible for **higher customer engagement and response**.  
-  Optimizing around these effective channels can enhance overall conversion rates.  
-  Channel-specific targeting should be considered in future campaigns.

---

### 5. 💰 **Annual Premium and Customer Tenure Insights**

- Customers with a **longer tenure (vintage)** show **slightly more interest** in renewing or purchasing insurance.  
-  This suggests a **loyalty effect** that can be nurtured.  
-  Introduce **loyalty rewards or tiered benefits** to encourage renewals.
